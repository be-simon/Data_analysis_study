```python
# 리스트를 전달하면 다수의 컬럼을 불러올 수 있음
df[["Symbol", "Name"]].sample(5)
```

```python
# 리스트를 전달하면 타입은 df
df[["Name"]].sample(5)
```

### loc로 sub set 가져오기

```python
# df.loc[raw, column]
df.loc[0,"Name"]
```

```python
# column index로 접근해도 같은 결과가 나오지만
# df["Name"] 의 결과로 Series를 리턴한 후 접근하기 때문에 불필요한 연산을 동반한다.
df["Name"][0]
```

```python
# 역시 리스트로 전달하면 여러 행, 열을 가져올 수 있다.
df.loc[[0,1,2], ["Name", "Symbol"]]
```

```python
# slice로 전달할 수도 있다.
df.loc[0:4, "Symbol":"Industry"]
```

### Boolean Indexing
* 결과를 T / F로 리턴
* 특정 조건을 만족하는 sub set 가져오기

```python
# df[조건식] -> 조건에 맞는 subset
df[df["Market"] == "KOSPI"].sample(5)
```

```python
# df.loc[조건식, 열]
# 반도체 관련 산업을 하는 기업들의 심볼:산업 가져오기
df.loc[df["Industry"].fillna("").str.contains("반도체") & (df["Market"] == 'KOSPI'), "Symbol":"Industry"]
```

### 파생변수
* 변수들을 연산해서 새로운 변수를 만든다

```python
# ListingDate 요소의 타입은 object
df[["ListingDate"]].dtypes
```

```python
df["ListingYear"] = pd.to_datetime(df["ListingDate"].fillna('1900')).dt.year
df[["Symbol", "Name", "ListingYear"]]
```
